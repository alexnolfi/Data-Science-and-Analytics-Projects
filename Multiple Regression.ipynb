```r
#Question 14

#Part A

set.seed(1)
x1=runif(100)
x2=0.5*x1+rnorm(100)/10
y=2+2*x1+0.3*x2+rnorm(100)
```

```r
The general form of this linear model is: Y = 2 + 2*X1 + 0.3*X2 + Error
The regression coefficients are B0=2, B1=2, and B2=0.3
```

```r
#Part B

cor(x1, x2)
```

```r
plot(x1, x2)
```

```r
#Part C

Multiple.Regression <- lm(y~x1+x2)
summary(Multiple.Regression)
```

```r
Based on the output of the regression, we can conclude that the variable x1
is statistically significant while the variable x2 is not. The coefficient
estimates are: B0=2.1305, B1=1.4396, and B2=1.0097. We can reject the null
hypothesis that B1=0 (since the result is statistically significant with a
p-value of 0.0487), but we cannot reject the null hypothesis that B2=0, as 
its p-value is far too high. In comparison to the population values of B0=2,
B1=2, and B2=0.3, the coefficient estimates are somwhat close (at 2.1305,
1.4396, and 1.0097), but with high standard errors.
```

```r
#Part D

Simple.Regression.X1 <- lm(y~x1)
summary(Simple.Regression.X1)
```

```r
Once again, we are able to reject the null hypothesis that B1=0. In fact,
the p-value is even lower in this case (at 2.66e-06), as the variable is now
the only predictor in the regression.
```

```r
#Part E

Simple.Regression.X2 <- lm(y~x2)
summary(Simple.Regression.X2)
```

```r
In this case, we can reject the null hypothesis that B2=0, as the p-value
is very low at 1.37e-05. At first glance, these results may appear
contradictory, but part F will explain why it is not a contradiciton.
```

```r
#Part F

Although the predictor x2 was found to be statistically significant in
part E, and not statistically significant in part C, this is not a contradiction.
When the multiple regression of y onto x1 and x2 is performed, it is difficult
to determine which of the two variables is more important in explaining y, as they
are collinear (with a correlation coefficient of 0.835). However, when the simple 
regressions are performed individually, both variables are able to adequately explain
the response. Thus, it is not surprising that when the simple regressions are
performed, we find that both variables are statistically significant.
```

```r
#Part G

#Adding the additional observation

x1=c(x1, 0.1)
x2=c(x2, 0.8)
y=c(y, 6)
```

```r
#Refitting Model c

Multiple.Regression.2 <- lm(y~x1+x2)
summary(Multiple.Regression.2)
```

```r
#Refitting Model d

Simple.Regression.1 <- lm(y~x1)
summary(Simple.Regression.1)
```

```r
#Refitting model e

Simple.Regression.2 <- lm(y~x2)
summary(Simple.Regression.2)
```

```r
The addition of the new observation has resulted in swapping the variables that
are statistically significant in the multiple regression model. Before the 
observation was added, x1 was statistically significant, and x2 was not statistically
significant. With the new observation, the opposite is true; x1 is not significant
and x2 is significant, with a p-value of 0.00614. In the simple regression models,
both of x1 and x2 are statistically significant in their respective models (as they
were before).
```

```r
#Identification of Outliers

To determine whether the new observation is an outlier in each model, 
we can use the "rstudent" command to compute the studentized residuals
for each model. Any values greater than absolute 3 are likely outliers.
```

```r
rstudent(Multiple.Regression.2) -> A; rstudent(Simple.Regression.1) -> B; 
rstudent(Simple.Regression.2) -> C
A[101]; B[101]; C[101]
```

```r
Based on the output above, it appears that the new observation is only
an outlier in the second model (Simple.Regression.1), as the 
studentized value of the new observation (the 101st observation in the 
list) is approximately 3.44. The other models have studentized residuals 
values well within the acceptable range.
```

```r
#Identification of High Leverage Points

To identify potential high leverage points, including the new observation, 
we can plot residuals from each of the models
```

```r
par(mfrow=c(2,2))
plot(Multiple.Regression.2)
plot(Simple.Regression.1)
plot(Simple.Regression.2)
```

```r
By examining the residuals vs leverage plot in each of the three models,
we can see that the 101st observation (i.e. the new observation) is a high
leverage point in the first and third models (the multiple regression and
the simple regression of y onto x2). This is the case because in both plots,
the 101st observation lies extremely far to the right from any neighbouring 
points (especially in the multiple regression). In no case, however, is the
new observation both a high leverage point and an outlier.
```
